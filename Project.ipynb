```python
import pandas as pd
#both zip file and csv files were present in same path that is why given only file's name not the whole path
df = pd.read_csv('orders.csv')
```

```python
df.head(20)
```

```python
#there are some 'not available', 'unknown' , 'None'"#Python code.py"
#we want to consider them as null

#to get distinct values available in a particular column
df['Ship Mode'].unique()
```

```python
# na_values -> treats particular value as null (converts)
df = pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

```python
df['Ship Mode'].unique()
```

```python
 #rename column names..make them lower case and replace space with underscore

df.rename(columns = {'Order Id':'order_id', 'City':'city','Order Date': 'order_date','Postal Code':'postal_code'})
```

```python
df.columns
```

```python
#can't pick cols one by one and change it, so use below method
df.columns = df.columns.str.lower()
```

```python
df.columns =df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

```python
#derive new columns discount, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']*.01
```

```python
df['discount'].head()
```

```python
df['sale_price'] = df['list_price'] - df['discount']
df
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
#convert order date from object data type to datetime -> it should go as a date not as string into database
#if it wont go as a date, we'll not be able to perform all the operations.
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format ="%Y-%m-%d")
```

```python
#drop cost price , list price  and discount percent columns - not required for analysis anymore
df.drop(columns=['list_price','cost_price','discount_percent'],inplace = True)
```
